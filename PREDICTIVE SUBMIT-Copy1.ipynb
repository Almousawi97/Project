```python
#code to split dataset, also used from https://github.com/ageron/handson-ml/blob/master/02_end_to_end_machine_learning_project.ipynb
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(df, 0.2)
print(len(train_set), "train +", len(test_set), "test")
```

To create a predictive model, both a test and training set are required. The training set is used to train the model so that it can learn how to predict accurately. Having trained the model, it is then tested again the test set to see how well it can make predictions based off of new pieces of data. 

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, link_match):
    ids = data[link_match]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
#seperating target variable from others 
X = df1.iloc[:, 0:25].values
y = df1.iloc[:, 25].values
```

```python
#splitting dataset for model testing with target variable
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=500, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

The first of the predictive models chosen and tested is the random forest model. This model will essentially combine different decision trees to predict the number of goals scored by the home team within any given game. After creating the model and training it, metrics were collected. Looking at the mean absolute error and the root mean squared error, it becomes clear that this model does need improvement. The mean absolute error is off by 0.8 goals and with an root mean squared error relatively far from the actual mean, improvements can be made.

```python
#running random forest model, used code from https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
#finding metrics on performance of random foorest model
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
